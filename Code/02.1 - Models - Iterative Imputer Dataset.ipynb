Using functions to pass data through a function transformer to allow their use in pipelines

```python
def get_scaled_columns(df):
    return pd.DataFrame(df[to_scale])
```

```python
def get_non_scaled_columns(df):
    return df[categoricals]
```

```python
df.columns[:15]
```

```python
df = df.drop(columns=['PassengerId', 'Cabin', 'Name', 'FirstName', 'CabinRoom'])
```

```python
to_scale = list(df.loc[:, 'Age':'TotalSpend'].columns)
to_scale
```

```python
categoricals = list(df.loc[:, 'HomePlanet_Earth':].columns)
categoricals
```

```python
get_scaled_transformer = FunctionTransformer(get_scaled_columns)
get_non_scaled_transformer = FunctionTransformer(get_non_scaled_columns)
```

```python
pipe_scaler =  Pipeline([
    ('scale_transform', get_scaled_transformer),
    ('ss', StandardScaler())
])

union = FeatureUnion([
    ('vect_pipe', pipe_scaler),
    ('other_cols', get_non_scaled_transformer)
])

union_res = union.fit_transform(df)
```

After thinking about it for a bit, it does make sense because with those categorical values represented as integers, they're being treated as ordinal values, and there really isn't a good way for the algorithm to effectively parse the values directly. This led me to learning about the concept of Target Statistics, which can be used in place of dimensionality bloating dummy encoding, where the category is given a value based on the likelihood of its effect on the target. I figure I was about half way to the concept here, but that last half would certainly have been a steep uphill climb. The next method I am going to use is the current state of the art for regression/classification models with categorical data, that leverages a special approach for generating the Target Statistics.

#### CatBoost Model

This model automatically generates target statistic values for each category, as well as combinations of categoricals (pretty much exactly what I was aiming to do with my earlier methods). As such, it will be fed the iterative imputer dataset without dummies or customization.

```python
import catboost
```

```python
df_ii = pd.read_csv('../Data/input_iterative_imputer.csv')
df_ii.head()
```

```python
df_ii.drop(['Cabin', 'Name', 'FirstName'], axis=1, inplace=True)
df_ii.head()
```

```python
id_associations = {}
for base_id in [id.split('_')[0] for id in df_ii['PassengerId']]:
    if base_id not in id_associations:
        id_associations[base_id] = 1
    else:
        id_associations[base_id] += 1
id_associations

# create feature to attribute this value
def get_assoc_value(row):
    return id_associations[row.split('_')[0]]

df_ii['Associates'] = df_ii['PassengerId'].apply(get_assoc_value)
```

```python
df_ii.head()
```

```python
df_ii['BaseId'] = df_ii['PassengerId'].str.split('_').str[0]
df_ii.head()
```

```python
cat_features = ['HomePlanet', 
                'CryoSleep', 
                'Destination', 
                'VIP',
#                 'LastName',
                'CabinSector',
                'CabinRoom',
                'CabinClass',
                'Associates',
#                 'BaseId',               
               ]
```

```python
X = df_ii.drop(['PassengerId', 'Transported', 'LastName', 'BaseId'], axis=1)
y = df_ii.Transported
```

```python
pool = catboost.Pool(data=X, label=y, cat_features=cat_features)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_validation, y_train, y_validation = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
train_pool = catboost.Pool(
    data=X_train,
    label=y_train,
    cat_features=cat_features
)

validation_pool = catboost.Pool(
    data=X_validation,
    label=y_validation,
    cat_features=cat_features
)
```

```python
from catboost import CatBoostClassifier, MetricVisualizer
```

```python
cb = CatBoostClassifier(
    iterations=1000,
    train_dir='Vanilla, 1000 iterations'
)

cb.fit(
    train_pool,
    eval_set=validation_pool,
    verbose=False,
    plot=True
);
```
